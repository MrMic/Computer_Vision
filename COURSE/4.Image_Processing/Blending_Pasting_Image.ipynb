## BLENDING IMAGES OF THE SAME SIZE

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
## !! Only works with images of THE SAME SIZES !! 
blended = cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.3,gamma=0)
```

```python
plt.imshow(blended)
```

## OVERLAY SMALL IMAGE ON TOP OF A LAYER IMAGE ( NO BLENDING )
## numpy reassignment

```python
img1 = cv2.imread('../../Computer-Vision-with-Python/DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('../../Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600,600))
```

```python
plt.imshow(img2)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
